```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "y1AQEIpMTR2j7xgr9MH0_Manufacturing Dataset.csv"  # Change this to your actual file path
df = pd.read_csv(file_path)

# Display the first 5 rows
print("\nðŸ”¹ First 5 Rows of the Dataset:")
print(df.head())

# Dataset shape
print(f"\nðŸ”¹ Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")

# Basic Info
print("\nðŸ”¹ Dataset Information:")
print(df.info())

# Summary Statistics
print("\nðŸ”¹ Summary Statistics:")
print(df.describe())

# Check for missing values
print("\nðŸ”¹ Missing Values in Dataset:")
print(df.isnull().sum())

# Handling Missing Values
df_cleaned = df.dropna()  # Drops rows with missing values (or use df.fillna() for imputation)
print(f"\nðŸ”¹ After Cleaning: {df_cleaned.shape[0]} rows remain.")

# Check for duplicate rows
print(f"\nðŸ”¹ Number of duplicate rows: {df.duplicated().sum()}")

# Visualizing Missing Values
plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull(), cmap="viridis", cbar=False, yticklabels=False)
plt.title("Missing Values Heatmap")
plt.show()

# Distribution of Numeric Features
df.hist(figsize=(12, 8), bins=30, edgecolor="black")
plt.suptitle("Feature Distributions")
plt.show()

# Correlation Matrix
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()

# Pairplot (for small datasets)
if df.shape[1] <= 5:  # Limit columns for better visualization
    sns.pairplot(df)
    plt.show()

# Countplot for Categorical Columns (if applicable)
categorical_columns = df.select_dtypes(include=["object"]).columns
for col in categorical_columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=df[col], palette="pastel", order=df[col].value_counts().index)
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()

print("\nâœ… EDA Completed!")
```
